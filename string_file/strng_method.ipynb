## **Slicing with Skip**

```python
print(l4[1:10])
print('--')
print(l4[1:10:2])
print('--')
print(l4[1:10:3])
```

```python
print(l2)
print('--')
print(l2[-1])
print('--')
print(l2[-2])
print('--')
print(l2[-5])
```

```python
print(l2[-2:-3]) # does not give output as it gives length 
```

```python
print(l2[-2:4])
print('--')
print(f"Prints complete list l2[::]= {l2[::]}")
print('--')
print(f"Reverse the list: {l2[::-1]}")
print('--')
print(f"Skips 2 steps: {l2[::2]}")
```
