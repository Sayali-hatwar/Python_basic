# Markdown types

``` Markdown ```

You're going to love using **Markdown** cells! They turn your Jupyter Notebook from a boring code file into a professional-looking report or presentation.

In Jupyter, change a cell from **Code** to **Markdown** using the dropdown menu at the top of the cell (or press `M` on your keyboard when the cell is selected).

Here is your "Cheat Sheet" for the most useful Markdown syntax:

---

### 1. Headers (The Titles)

Use the `#` symbol. The more hashes you use, the smaller the header.

```markdown
# Level 1 Header (Main Title)
## Level 2 Header (Section)
### Level 3 Header (Sub-section)

```

### 2. Text Styling

Make your insights stand out with bold or italics.

* **Bold**: `**Bold Text**` or `__Bold Text__`
* *Italic*: `*Italic Text*` or `_Italic Text_`
* ~~Strikethrough~~: `~~Strikethrough~~`

### 3. Lists (Bullet Points)

Great for listing steps or data features.

* **Unordered**: Use `*`, `-`, or `+`
* Item 1
* Item 2


* **Ordered**: Use numbers
1. First Step
2. Second Step



### 4. Mathematical Equations (LaTeX)

Since you are likely doing data science, you’ll need math. Jupyter uses LaTeX for this.

* **Inline**: Enclose in single dollar signs, like `$E = mc^2$`.
* **Display (Centered)**: Enclose in double dollar signs:

$$\frac{a}{b} = c$$



### 5. Links and Images

* **Link**: `[Clickable Text](https://www.google.com)`
* **Image**: `![Alt Text](url_to_image.jpg)`

### 6. Blockquotes & Code Snippets

If you want to highlight a note or show a piece of code without running it:

> **Note:** This is a blockquote for important information.

To show code:
`     ` `python print("This code won't run, it's just for display") ` `     `

---

### Pro Tip: The "Table" Shortcut

Tables are great for summarizing data manually.

```markdown
| Name | Age | Department |
|------|-----|------------|
| Sayali| 25  | Analytics  |
| Gemini| ??  | AI         |

```
